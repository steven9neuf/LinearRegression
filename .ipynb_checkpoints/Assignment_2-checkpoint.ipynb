3- Create a new column called 'Enroll_Rate' and which must contain the % proportion of applicants enrolled over the accepted ones.
Which college has the maximum Enroll_Rate ? Which one has the minimum ? 

```python
# your code here
df['Enroll_Rate'] = round(df.Enroll / df.Accept ,4)
name = df['Enroll_Rate'].argmax()
rate = df['Enroll_Rate'].loc[df['Enroll_Rate'].argmax()]
print('{} has the maximum Enroll_Rate with {}%'.format(name,rate*100))

name = df['Enroll_Rate'].argmin()
rate = df['Enroll_Rate'].loc[df['Enroll_Rate'].argmin()]
print('{} has the minimum Enroll_Rate with {}%\n'.format(name,rate*100))
```

## In the following, you are going to build a Linear Regression model and then use it to predict the number of applications received as a fucntion of all the features present in the College Dataset


First, let's plot the Target variable as a fucntion of each feature (the code is gently provided for you:)) 

```python
import matplotlib.pyplot as plt
%matplotlib inline
for col in df.columns:
    plt.scatter(df[col], df.Apps,)
    plt.title('Nb applicants VS '+ col)
    plt.xlabel(col)
    plt.ylabel('Nb applicants')
    plt.show()
```

4- Load the target variable "Apps" into a dataframe and call it y. Then, load the features into a second dataframe and call it X.

```python
# your code here
y = df.Apps
X = df.drop('Apps',axis=1,inplace=False)
#print(y,'\n',X)
```

5- Plot a histogram of y, choose the number of bins as 100.

```python
%matplotlib inline 
#The "inline" enables plotting figres directly within the notebook
# your code here
fig = plt.figure(figsize=(10, 5))
fig.clf()
ax = fig.gca()
y.hist(bins = 100, ax = ax)
ax.set_xlabel('Applications')
ax.set_ylabel('Density of applications')
ax.set_title('Density of applications') 
```

6- Split the data set into a training set and a test set. Set: test_size=0.3, random_state=123

```python
# your code here
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
print ('training test : ', X_train.shape,y_train.shape)
print ('testing test : ', X_test.shape,y_test.shape)
```

7- Fit a linear model to the training set, and then report the training and testing errors obtained (the R2 statistic)

```python
# your code here
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
#print(reg)
reg.fit(X_train, y_train)
train_score = reg.score(X_train, y_train)
test_score = reg.score(X_test, y_test)
print  ('train score =' ,  train_score)
print  ('test score = ', test_score)
```

9- When you look back at the histogram of y, there is a single value to the right that is far from the others. This might be an outlier (abnormal data point). <br>
Remove this maximum value from df

```python
# your code here
df.drop(df.Apps.argmax(),axis=0,inplace=True)
```

10- Now, having removed the outlier, Re-train your linear model with the new dataframes y, X. 
Print the train/test score. Did these scores improve with respect to earlier ones?

```python
# Your code here
y = df.Apps
X = df.drop('Apps',axis=1,inplace=False)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

reg = LinearRegression()
#print(reg)
reg.fit(X_train, y_train)
train_score = reg.score(X_train, y_train)
test_score = reg.score(X_test, y_test)
print  ('train score =' ,  train_score)
print  ('test score = ', test_score)
```
